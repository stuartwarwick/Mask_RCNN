# Sampling honeycomb images

```python
WORKING_DIRECTORY = 'C:\\Users\\Lenovo ThinkPad\My Drive\\bee images OX RSousa\\apidea frames brood assess'

import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from icecream import ic
from PIL import Image
```

```python
# Loop all Camera1 video files to produce the subtraction optimised meniscus tracking info 
file_regex = WORKING_DIRECTORY + '/*.J*'

for f in glob.glob(file_regex):
    filename = f.split('\\')[-1]
    ic('Processing', filename)
        
    # Open the input video file
    with Image.open(f) as im:
        w,h = im.size
        (left, upper, right, lower) = (int(w/3), int(2*h/3), int(2*w/3), h-20)
        im_crop = im.crop((left, upper, right, lower))
        im_crop.save(filename)
        im_crop.close()
    
    
```
